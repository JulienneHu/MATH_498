```r
library(tidyverse)
library(ggplot2)
library(MASS)
```

```r
df <- read.csv("2309_train.csv")[,-1]
df_initial <- df
```

```r
FILTER <- function(x) {
    x1 <- x
    for(i in 1:9) {
        x2 <- x1 %>% dplyr::filter((x1[,i] <= (mean(x1[,i])+3*sd(x1[,i]))) & (x1[,i] >= (mean(x1[,i])-3*sd(x1[,i]))))
        x1 <- x2
    }
    return (x1)
}
```

```r
dff <- FILTER(df)
```

```r
dff$X6 <- factor(dff$X6)
dff$X8 <- factor(dff$X8)
set.seed(12345)
test_size <- floor(nrow(dff) * 0.3)
test_id <- sample(1:nrow(dff), test_size)
testdff <- dff[test_id,]
traindff <- dff[-test_id,]
```

```r
mod1 <- lm(Y1 ~ X1 + X2 + X3  + X5 + X6 + X7 + X8 + X9, data = traindff)
summary(mod1)
```

```r
train_predict1 <- predict.lm(mod1, traindff)
(train_mse_OLS1 <- mean((train_predict1 - traindff$Y1)^2))

test_predict1 = predict.lm(mod1, testdff)
(test_mse_OLS1 = mean((test_predict1 - testdff$Y1)^2))
```

```r
mod2 <- lm(Y1 ~ X1 + X2 + X3  + X5 + X6 + X7 + X8 + X9 + I(X1^2) + I(X2^2) + I(X3^2) + 
           I(X5^2) + I(X7^2) + I(X9^2) + I(X1^3) + I(X2^3) + I(X3^3) + I(X5^3) +
           I(X7^3) + I(X9^3) + I(X1^4) + I(X2^4) + I(X3^4) + I(X5^4) +
           I(X7^4) + I(X9^4) + 
           X1*X2 + X1*X3 + X1*X5 + X1*X6 + X1*X7 + X1*X8 + X1*X9 + 
           X2*X3 + X2*X5 + X2*X6 + X2*X7 + X2*X8 + X2*X9 + 
           X3*X5 + X3*X6 + X3*X7 + X3*X8 + X3*X9 + 
           X5*X6 + X5*X7 + X5*X8 + X5*X9 + 
           X6*X7 + X6*X8 + X6*X9 +
           X7*X8 + X7*X9 +
           X8*X9 ,
           data = traindff)
```
